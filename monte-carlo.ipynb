```python
import random
```

```python
N=[100, 500, 1000, 5000, 10000, 50000, 100000]
inCircle=0
pi_estimates=[]
```

```python
for i in range(len(N)):
    for j in range(N[i]):
        x=random.uniform(-1,1)
        y=random.uniform(-1,1)
        if (pow(x,2)+pow(y,2)<=1):
            inCircle+=1
    pi=4*inCircle/N[i]
    pi_estimates.append(pi)
    inCircle=0
```

```python
for i in range(len(pi_estimates)):
    print(N[i],'Points: ', pi_estimates[i])
```
